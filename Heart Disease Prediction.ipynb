```python
from sklearn.preprocessing import StandardScaler
categorical=['sex','cp','fbs','restecg','exang','ca','thal']
data=pd.get_dummies(data,columns=categorical,drop_first=True)
numeric=['age','trestbps','chol','thalach','oldpeak']
scaler=StandardScaler()
data[numeric]=scaler.fit_transform(data[numeric])
```

```python
data['age_chol']=data['age']*data['chol']
for c in numeric:
    q1=data[c].quantile(0.25)
    q3=data[c].quantile(0.75)
    iqr=q3-q1
    low=q1-1.5*iqr
    high=q3+1.5*iqr
    data[c]=data[c].clip(low, high)
print(data.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import f1_score, roc_auc_score, confusion_matrix
X=data.drop("condition",axis=1)
y=data["condition"]
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
model1=LogisticRegression(max_iter=1000)
model1.fit(X_train,y_train)
pred1=model1.predict(X_test)
```

```python
model2=DecisionTreeClassifier()
model2.fit(X_train,y_train)
pred2=model2.predict(X_test)
```

```python
print("Logistic Regression:")
print("F1:",f1_score(y_test,pred1))
print("AUC:",roc_auc_score(y_test,model1.predict_proba(X_test)[:,1]))
print("Confusion Matrix:\n",confusion_matrix(y_test,pred1))
```

```python
print("\nDecision Tree")
print("F1:",f1_score(y_test,pred2))
print("AUC:",roc_auc_score(y_test,model2.predict_proba(X_test)[:,1]))
print("Confusion Matrix:\n",confusion_matrix(y_test, pred2))
```
